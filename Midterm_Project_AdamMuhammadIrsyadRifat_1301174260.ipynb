#### Code

```python
data1 = pd.read_csv(io.BytesIO(uploaded['hang_seng.csv'])) #WARNING - pastikan nama csv dalam code sama dengan yang di upload (huruf kapital berlaku)
data2 = pd.read_csv(io.BytesIO(uploaded['nasdaq.csv'])) #WARNING - pastikan nama csv dalam code sama dengan yang di upload (huruf kapital berlaku)
data3 = pd.read_csv(io.BytesIO(uploaded['nikkei.csv'])) #WARNING - pastikan nama csv dalam code sama dengan yang di upload (huruf kapital berlaku)
```

```python
#menyatukan file csv 1,2,3
#frames = [data1, data2, data3] 
#result = pd.concat(frames).reset_index(drop=True)
#result.to_csv('hnn.csv', index=False)
```

```python
import datetime

hnn1_df = data1
hnn2_df = data2
hnn3_df = data3 
#hnn1_df.head()
#hnn2_df.head()
#hnn3_df.head()

#csv hang_seng(1)
hnn1_df = hnn1_df.round(2)
#hnn1_df.head()
hnn1_df.isnull().sum()
hnn1_df.dropna(inplace = True, axis = 0)
hnn1_df['Date'] = pd.to_datetime(hnn1_df['Date'])
#hnn1_df.head(2)
hnn1_df['Date'].max() - hnn1_df['Date'].min()
#hnn1_df.iloc[-90:].describe().astype(int)

#csv nasdaq(2)
hnn2_df = hnn2_df.round(2)
hnn2_df.isnull().sum()
hnn2_df.dropna(inplace = True, axis = 0)
#hnn2_df.head()
hnn2_df['Date'] = pd.to_datetime(hnn2_df['Date'])
#hnn2_df.head(2)
hnn2_df['Date'].max() - hnn2_df['Date'].min()
#hnn2_df.iloc[-90:].describe().astype(int)

#csv nikkei (3)
hnn3_df = hnn3_df.round(2)
hnn3_df.isnull().sum()
hnn3_df.dropna(inplace = True, axis = 0)
#hnn3_df.head()
hnn3_df['Date'] = pd.to_datetime(hnn3_df['Date'])
#hnn3_df.head(2)
hnn3_df['Date'].max() - hnn3_df['Date'].min()
#hnn3_df.iloc[-90:].describe().astype(int)
```

```python
#1
hnn1_df.index = hnn1_df['Date']
hnn1_sele = hnn1_df.loc['2019-1-1':'2019-11-30']
hnn1_sele['Adj Close'].plot(figsize = (8,4))
plt.show()
#2
hnn2_df.index = hnn2_df['Date']
hnn2_sele = hnn2_df.loc['2019-1-1':'2019-11-30']
hnn2_sele['Adj Close'].plot(figsize = (8,4))
plt.show()
#3
hnn3_df.index = hnn3_df['Date']
hnn3_sele = hnn3_df.loc['2019-1-1':'2019-11-30']
hnn3_sele['Adj Close'].plot(figsize = (8,4))
plt.show()
```

```python
hnn1_sele['Day_Perc_Change'] = hnn1_df['Adj Close'].pct_change()*100
hnn1_sele['Day_Perc_Change'].hist(bins = 50, figsize = (10,5)) 
hnn2_sele['Day_Perc_Change'] = hnn2_df['Adj Close'].pct_change()*100
hnn2_sele['Day_Perc_Change'].hist(bins = 50, figsize = (10,5)) 
hnn3_sele['Day_Perc_Change'] = hnn3_df['Adj Close'].pct_change()*100
hnn3_sele['Day_Perc_Change'].hist(bins = 50, figsize = (10,5)) 
plt.xlabel('Daily returns')
plt.ylabel('Frequency')
plt.legend(labels='nikkei', loc='best')
plt.show()
```

```python
def trend(x):
  if x > -0.5 and x <= 0.5:
    return 'Slight or No change'
  elif x > 0.5 and x <= 1:
    return 'Slight Positive'
  elif x > -1 and x <= -0.5:
    return 'Slight Negative'
  elif x > 1 and x <= 3:
    return 'Positive'
  elif x > -3 and x <= -1:
    return 'Negative'
  elif x > 3 and x <= 7:
    return 'Among top gainers'
  elif x > -7 and x <= -3:
    return 'Among top losers'
  elif x > 7:
    return 'Bull run'
  elif x <= -7:
    return 'Bear drop'

#1 (hang_seng)   
hnn1_sele['Trend']= np.zeros(hnn1_sele['Day_Perc_Change'].count())
hnn1_sele['Trend']= hnn1_sele['Day_Perc_Change'].apply(lambda x:trend(x))
hnn1_sele.head()

hnn1_pie_data = hnn1_sele.groupby('Trend')
pie_label = sorted([i for i in hnn1_sele.loc[:, 'Trend'].unique()])
plt.pie(hnn1_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

```python
#2   (nasdaq)
hnn2_sele['Trend']= np.zeros(hnn2_sele['Day_Perc_Change'].count())
hnn2_sele['Trend']= hnn2_sele['Day_Perc_Change'].apply(lambda x:trend(x))
hnn2_sele.head()

hnn2_pie_data = hnn2_sele.groupby('Trend')
pie_label = sorted([i for i in hnn2_sele.loc[:, 'Trend'].unique()])
plt.pie(hnn2_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

```python
#3  (nikkei) 
hnn3_sele['Trend']= np.zeros(hnn3_sele['Day_Perc_Change'].count())
hnn3_sele['Trend']= hnn3_sele['Day_Perc_Change'].apply(lambda x:trend(x))
hnn3_sele.head()

hnn3_pie_data = hnn3_sele.groupby('Trend')
pie_label = sorted([i for i in hnn3_sele.loc[:, 'Trend'].unique()])
plt.pie(hnn3_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# type your code here
#hang_Seng(1) warna biru
hnn1_sele['Day_Perc_Change'] = hnn1_df['Adj Close'].rolling(7).std()*np.sqrt(7)
hnn1_sele.dropna(axis = 0, inplace = True)
hnn1_sele['Day_Perc_Change'].plot(figsize = (28, 8), fontsize = 11)
plt.legend(labels='hang_seng', loc='best') 
#nasdaq(2)warna orange
hnn2_sele['Day_Perc_Change'] = hnn2_df['Adj Close'].rolling(7).std()*np.sqrt(7)
hnn2_sele.dropna(axis = 0, inplace = True)
hnn2_sele['Day_Perc_Change'].plot(figsize = (28, 8), fontsize = 11)
plt.legend(labels='nasdaq', loc='best')
#nikkei(3) warna hijau
hnn3_sele['Day_Perc_Change'] = hnn3_df['Adj Close'].rolling(7).std()*np.sqrt(7)
hnn3_sele.dropna(axis = 0, inplace = True)
hnn3_sele['Day_Perc_Change'].plot(figsize = (28, 8), fontsize = 11)
plt.legend(labels='nikkei', loc='best')
```
